```python
import numpy as np
from numpy import linalg
import scipy
```

```python
n = 10
np.random.seed(seed = 42)
A = np.random.randn(n,n)

x_true = np.ones(n)

x_true = x_true.T

b = A @ x_true
print(b)
```

```python
#2 - norm
print(f"Cond number using 2-norm: {np.linalg.cond(A,2)} \n")

# inf - norm
print(f"Cond number inf norm: {np.linalg.cond(A,np.inf)} \n")
```

The higher the condition number, the worse the accuracy of the solution (higher error). 
We choose the threshold 1e5 to determine as a testing criterion if the matrix is ill-conditioned.
as we can observe, both the 2-norm and the INF-norm are much smaller than the threshold.
=> the matrix is well conditioned

```python
# Solve the linear system Ax = b with the function np.linalg.solve().

x_sol = np.linalg.solve(A,b)
print(x_true)
print(x_sol)
```

```python
# Computes the relative error between the solution computed before and the true solution xtrue.
#Remember that the relative error between xtrue and x in Rn can be computed as

E_rel = np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2)
print(f"The relative error is {E_rel}")
```

```python
#Plot a graph (using matplotlib.pyplot) with the relative errors as a function of n and (in a new
# window) the condition number in 2-norm K2(A) and in ∞-norm, as a function of n.

#1e
maxRange = 100
norm2 = []
norm_inf = []
E_rel = []
for n in range(1,maxRange):
    A = np.random.randn(n, n)
    x_true = np.ones(n).T
    b = A @ x_true
    x_sol = np.linalg.solve(A, b)
    
    norm2.append(np.linalg.cond(A, 2))
    norm_inf.append(np.linalg.cond(A, np.inf))
    E_rel.append(np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2))
```

```python
import matplotlib.pyplot as plt

x_axis = range(1,maxRange)

plt.figure(figsize=(20,10))
plt.suptitle("Relative error and condition number")
plt.subplot(1,2,1)
plt.plot(x_axis, E_rel, color="red")
plt.grid()
plt.xlabel("Number of dimensions")
plt.ylabel("Relative error")

plt.subplot(1,2,2)
plt.plot(x_axis, norm2)
plt.plot(x_axis, norm_inf)
plt.grid()
plt.xlabel("Number of dimensions")
plt.ylabel("Condition number")
plt.legend(["2-norm", "inf-norm"])

plt.show()
```

```python
#first bulet point - random matrix
n_vals = np.arange(1,101,10)
norm2 = []
norm_inf = []
E_rel = []
for n in n_vals:
    A = np.random.randn(n, n)
    x_true = np.ones(n).T
    b = A @ x_true
    x_sol = np.linalg.solve(A, b)
    
    norm2.append(np.linalg.cond(A, 2))
    norm_inf.append(np.linalg.cond(A, np.inf))
    E_rel.append(np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2))

x_axis = range(1,maxRange)

plt.figure(figsize=(20,10))
plt.suptitle("Relative error and condition number for a random matrix")
plt.subplot(1,2,1)
plt.plot(n_vals, E_rel, color="red")
plt.grid()
plt.xlabel("Number of dimensions")
plt.ylabel("Relative error")

plt.subplot(1,2,2)
plt.plot(n_vals, norm2)
plt.plot(n_vals, norm_inf)
plt.grid()
plt.xlabel("Number of dimensions")
plt.ylabel("Condition number")
plt.legend(["2-norm", "inf-norm"])

plt.show()

###


# Second buletpoint - Vander matrix
n_vals = np.arange(5,31,5)
norm2 = []
norm_inf = []
E_rel = []
for n in n_vals:
    n = int(n)
    x = range(1,n + 1)
    A = np.vander(x, n)
    
    x_true = np.ones(n).T
    b = A @ x_true
    x_sol = np.linalg.solve(A, b)
    
    norm2.append(np.linalg.cond(A, 2))
    norm_inf.append(np.linalg.cond(A, np.inf))
    E_rel.append(np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2))


plt.figure(figsize=(20,10))
plt.suptitle("Relative error and condition number for Vander matrix ")
plt.subplot(1,2,1)
plt.plot(n_vals, E_rel, color="red")
plt.grid()
plt.xlabel("Number of dimensions")
plt.ylabel("Relative error")

plt.subplot(1,2,2)
plt.plot(n_vals, norm2)
plt.plot(n_vals, norm_inf)
plt.grid()
plt.xlabel("Number of dimensions")
plt.ylabel("Condition number")
plt.legend(["2-norm", "inf-norm"])

plt.show()


###

# Third buletpoint - Hilbert matrix
n_vals = np.arange(4,14,1)
norm2 = []
norm_inf = []
E_rel = []
for n in n_vals:
    n = int(n)
    A = scipy.linalg.hilbert(n)
    x_true = np.ones(n).T
    b = A @ x_true
    x_sol = np.linalg.solve(A, b)
    
    norm2.append(np.linalg.cond(A, 2))
    norm_inf.append(np.linalg.cond(A, np.inf))
    E_rel.append(np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2))


plt.figure(figsize=(20,10))
plt.suptitle("Relative error and condition number for Hilbert matrix")
plt.subplot(1,2,1)
plt.plot(n_vals, E_rel, color="red")
plt.grid()
plt.xlabel("Number of dimensions")
plt.ylabel("Relative error")

plt.subplot(1,2,2)
plt.plot(n_vals, norm2)
plt.plot(n_vals, norm_inf)
plt.grid()
plt.xlabel("Number of dimensions")
plt.ylabel("Condition number")
plt.legend(["2-norm", "inf-norm"])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Create an array of n values. This range is more reasonable for floating point calculations.
n_vals = np.arange(1, 101, 1)
err = []
computed_values = []

# Compute the error for each n
for n in n_vals:
    an = (1 + 1/n)**n
    err.append(abs(math.exp(1) - an))
    computed_values.append(an)



plt.figure(figsize=(15,6))
plt.suptitle("Approximation of the Euler constant (with relative error)")
plt.subplot(1,2,1)
plt.plot(n_vals,computed_values)
plt.plot(n_vals,[np.e]*100,"red")
plt.legend(["Calculated","True value"])
plt.ylabel("Value")
plt.xlabel("Value of n")

# Plotting the error
plt.subplot(1,2,2)
plt.plot(n_vals, err, 'r')
plt.xlabel('n')
plt.ylabel('Relative Error')
plt.title('Error in approximation of e')
plt.show()
```
